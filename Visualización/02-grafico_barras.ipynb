```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(movies)), num_oscars)
plt.title("Películas más premiadas")
plt.ylabel("Nº de premios de la academia")
plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
plt.savefig("películas.png")
plt.gca().clear()
```

```python
#Crear un historgrama con los siguientes datos

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
from collections import Counter
bar_size = 10 #tamaño de los intervalos
histogram = Counter(min(grade // bar_size * bar_size, 100 - bar_size) for grade in grades)
histogram
```

```python
plt.bar([x+bar_size/2 for x in histogram.keys()], 
         histogram.values(), 
         bar_size, 
         edgecolor = (0, 0, 0)
       )
plt.axis([-bar_size/2, 100 + bar_size/2, 0, 5])
plt.xticks([bar_size * i for i in range(100//bar_size + 1)])
plt.title("Distribución resultados del examen")
plt.ylabel("Nº estudiantes")
plt.show()
```

```python
plt.savefig("Resultados examen.png")
```
